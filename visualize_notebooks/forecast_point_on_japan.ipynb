```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pathlib

import geopandas as gpd
import seaborn as sns
sns.set()

ROOT_PATH = pathlib.Path().resolve().parents[0]
plt.rcParams["font.family"] = "Yu Gothic"
```

```python
japan = gpd.read_file(ROOT_PATH/"shp/japank.shp")
```

```python
df = pd.read_csv(ROOT_PATH/"csv/2022-01-01_msm_data.csv")
```

```python
fig,ax = plt.subplots()
ax.set_facecolor("lightgray")
japan.plot(ax=ax,color='#90ee90',edgecolor="black",linewidth=0.2)

for i in df.iterrows():
    ax.plot(i[1]["lon"],i[1]["lat"],color="red",marker="o",markersize=2)

ax.set_xlim([122,150])
ax.set_title("()")
ax.set_ylabel("lat")
ax.set_xlabel("lon")
plt.savefig(ROOT_PATH/"png/forecast_point.png",facecolor="white")
plt.show()
```
