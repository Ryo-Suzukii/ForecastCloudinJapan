```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import pathlib
from sklearn.model_selection import train_test_split
import pickle

ROOT_PATH = pathlib.Path().resolve().parents[0]
```

```python
X_df = pd.read_csv(ROOT_PATH/"csv/2023-07-03X_df.csv")
y_df = pd.read_csv(ROOT_PATH/"csv/2023-07-03y_df.csv")["ncld"]
```

```python
X_df["time"] = pd.to_datetime(X_df["time"])
X_df = X_df.drop(["time","lat","lon"],axis=1)
```

```python
X_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3)
lgb_train = lgb.Dataset(X_train,y_train)
```

```python
params = {
    'objective': 'regression',
    'metric': 'mse'
}
```

```python
model = lgb.train(params, lgb_train)
```

```python
filename = "models/model_using_msm_extract_only_amedas_point.pkl"
with open(ROOT_PATH/filename,"wb") as f:
    pickle.dump(model,f)
```
