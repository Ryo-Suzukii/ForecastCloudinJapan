```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import japanize_matplotlib
import datetime as dt

from sklearn.model_selection import train_test_split
import lightgbm as lgb

import sys
sys.path.append("../")

import modules
```

```python
date = dt.date(2022,1,1)
datasets = []

while True:
    try:
        tmp_ds = xr.open_dataset(f"../data/{date.year}/{str(date.month).zfill(2)}/{date.year}_{str(date.month).zfill(2)}{str(date.day).zfill(2)}.nc")
        datasets.append(tmp_ds)
    except:
        print(f"except in:{date}")

    date += dt.timedelta(days=1)

    if date.year != 2022:
        break

    # debug
    if date.month > 1:
        break
    
    print(f"\r{date}",end="")
combined_dataset = xr.concat(datasets,dim="time")
```
